```python
from sklearn import datasets
import matplotlib.pyplot as plt

wines = datasets.load_wine()
wines_X = wines.data[:, :2]
wines_y = wines.target

x_min, x_max = wines_X[:, 0].min() - .5, wines_X[:, 0].max() + .5
y_min, y_max = wines_X[:, 1].min() - .5, wines_X[:, 1].max() + .5

plt.figure(2, figsize=(8, 6))
plt.clf()
# Plot the training points
plt.scatter(wines_X[:, 0], wines_X[:, 1], c=wines_y, cmap=plt.cm.Paired)
plt.xlabel('alcohol')
plt.ylabel('Smalic_acid')

plt.xlim(x_min, x_max)
plt.ylim(y_min, y_max)
plt.xticks(())
plt.yticks(())
```

```python
from sklearn.decomposition import PCA

X_reduced = PCA(n_components=3).fit_transform(wines.data)

x_min, x_max = X_reduced[:, 0].min() - .5, X_reduced[:, 0].max() + .5
y_min, y_max = X_reduced[:, 1].min() - .5, X_reduced[:, 1].max() + .5

plt.figure(2, figsize=(8, 6))
plt.clf()
# Plot the training points
plt.scatter(X_reduced[:, 0], X_reduced[:, 1], c=wines_y, cmap=plt.cm.Paired)
plt.xlabel('PC1')
plt.ylabel('PC2')

plt.xlim(x_min, x_max)
plt.ylim(y_min, y_max)
plt.xticks(())
plt.yticks(())
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression().fit(X_reduced, wines_y)
```
